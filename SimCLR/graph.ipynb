```python
import matplotlib.pyplot as plt

# Data for SimCLR (original)
batch_sizes = [32, 64, 128]
accuracies_simclr = [78.46, 80.23, 80.84]

# Data for BYOL
accuracies_byol = [76.42, 78.23, 79.84]

# Data for DINO (new: x=32: 77.65, x=64: 79.18, x=128: 80.5)
accuracies_dino = [77.65, 79.18, 80.5]

# Reverse for plotting: largest batch size on the left
batch_sizes_rev = batch_sizes[::-1]      # [128, 64, 32]
accuracies_simclr_rev = accuracies_simclr[::-1]  # [80.84, 80.23, 78.46]
accuracies_byol_rev = accuracies_byol[::-1]      # [79.84, 78.23, 76.42]
accuracies_dino_rev = accuracies_dino[::-1]      # [80.5, 79.18, 77.65]

# Create a 4x4 inch square figure with white background
fig = plt.figure(figsize=(4, 4), facecolor='white')
ax = fig.add_subplot(111)

# Plot SimCLR: red line with filled circles
ax.plot(batch_sizes_rev, accuracies_simclr_rev,
        marker='s', markersize=10, markerfacecolor='red', markeredgecolor='red',
        color='red', linewidth=3, label='SimCLR')

# Plot BYOL: green line with filled squares
ax.plot(batch_sizes_rev, accuracies_byol_rev,
        marker='o', markersize=10, markerfacecolor='green', markeredgecolor='green',
        color='green', linewidth=3, label='BYOL')

# Plot DINO: blue line with filled diamonds
ax.plot(batch_sizes_rev, accuracies_dino_rev,
        marker='D', markersize=10, markerfacecolor='blue', markeredgecolor='blue',
        color='blue', linewidth=3, label='DINO')

# Grid
ax.grid(True, linestyle='--', alpha=0.7, zorder=0)

# Labels and title (title commented out as before)
ax.set_xlabel('Batch size', fontsize=14, labelpad=10)
ax.set_ylabel('k-NN Top-1 Accuracy (%)', fontsize=14, labelpad=10)
# ax.set_title('(a) Impact of batch size', fontsize=16, pad=20)

# X-axis: exact ticks
ax.set_xticks(batch_sizes_rev)
ax.set_xticklabels(batch_sizes_rev, fontsize=12)

# Y-axis: extended to comfortably fit the new highest point (80.5)
ax.set_ylim(76.0, 81.2)
ax.set_yticks([76.0, 76.5, 77.0, 77.5, 78.0, 78.5, 79.0, 79.5, 80.0, 80.5, 81.0])
ax.tick_params(axis='y', labelsize=12)

# Invert x-axis so larger batch size is on the left
ax.invert_xaxis()

# Legend: with three lines rising to the left, 'lower left' avoids overlap with data points
ax.legend(fontsize=12, loc='lower left', frameon=True, fancybox=False, edgecolor='black')

# Tight layout with original padding
fig.tight_layout()
fig.subplots_adjust(left=0.18, right=0.85, top=0.82, bottom=0.18)

fig = plt.figure(figsize=(10, 10), facecolor='white')
# Save the plot
plt.savefig('ablation_bs.pdf', bbox_inches='tight')
```
